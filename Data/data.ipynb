```python
import pandas as pd

# Step 1: Read the CSV file into a DataFrame
pd.read_csv('IRELAND_CRIME_GARDA_DIVISION_wise_2003-2019.csv', header = None).to_csv('file3.csv', header = False)
```

```python
import csv

input_file = "file3.csv"
output_file = "crime_normalized.csv"

with open(input_file, newline='', encoding='utf-8') as infile, \
     open(output_file, 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = [
        "id", "region", "garda_division",
        "offence_code", "offence", "type_of_offence",
        "year", "quarter", "offence_count"
    ]
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    i = 0
    for row in reader:
        for key in row:
            if "Q" in key:  # identifies columns like 2003Q1
                year = int(key[:4])
                quarter = int(key[-1])
                i = i+ 1
    
                writer.writerow({
                    "id": i,
                    "region": row["REGION"],
                    "garda_division": row["GARDA DIVISION"],
                    "offence_code": row["OFFENCE CODE"],
                    "offence": row["OFFENCE"],
                    "type_of_offence": row["TYPE_OF_OFFENCE"],
                    "year": year,
                    "quarter": quarter,
                    "offence_count": int(row[key]) if row[key] else 0
                })
```
